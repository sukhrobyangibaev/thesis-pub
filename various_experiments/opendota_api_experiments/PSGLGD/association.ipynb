```python
import matplotlib.pyplot as plt
import pandas as pd
dataset = pd.read_csv('psglgd.csv').iloc[:, 1:]
dataset.head()
```

```python
pd.DataFrame({'columns': dataset.columns.tolist()})
```

```python
dataset = dataset.iloc[:, 4:14]
```

```python
dataset
```

```python
heroes = []
for i in range(0, 1948):
    heroes.append([str(dataset.values[i, j]) for j in range(0, 10)])
```

```python
heroes[0]
```

```python
from apyori import apriori
rules = apriori(transactions=heroes, min_support=0.003, min_confidence=0.2, min_lift=3, max_length=2)
```

```python
results = list(rules)
results
```

```python
def inspect(results):
    lhs = [tuple(result[2][0][0])[0] for result in results]
    rhs = [tuple(result[2][0][1])[0] for result in results]
    supports = [result[1] for result in results]
    confidence = [result[2][0][2] for result in results]
    lifts = [result[2][0][3] for result in results]
    return list(zip(lhs, rhs, supports, confidence, lifts))
resultsInDataFrame = pd.DataFrame(inspect(results), columns=['Left Hand Side', 'Right Hand Side', 'Support', 'Confidence', 'Lift'])
```
